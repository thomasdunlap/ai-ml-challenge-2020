# ImageMagick Converts PDF to JPG

![Image of ImageMagick Wizard]('../app/static/ImageMagick')

https://imagemagick.org/index.php

ImageMagick is a free Opensource tool for modifying image files.  We use this to convert the PDF document to a JPG, so Tesseract can use OCR.  OCR requires an image file, and PDFs are a type of binary file, which is why we convert them to a JPG.  We use the internal Python module subprocess to call the Imagemagick program's `convert` function for creating the JPG output image.



```python
import subprocess


def convert_to_image(in_file, out_file_prefix):
    
    cmd = "convert -density 300 {} {}.jpg".format(in_file, out_file_prefix)
    return subprocess.call(cmd, shell=True)

convert_to_image('../app/static/uploads/sample_eula_2.pdf', '../app/static/uploads/sample_eula_2')
```

# Google Tesseract OCR

From version 4 onwards, Google has given a significant boost to this OCR engine. Tesseract 4.0 has added a new OCR engine that uses a neural network system based on LSTM (Long Short-term Memory), one of the most effective solutions for sequence prediction problems. Although its previous OCR engine using pattern matching is still available as legacy code.

Once you have downloaded Tesseract onto your system, you easily run it from the command line using the following command:

`tesseract <test_image> <output_file_name> -l <language(s)> --oem <mode> --psm <mode>`

You can also use pytesseract, a Python-wrapped version

```python
im = cv2.imread('../app/static/uploads/sample_eula_2-2.jpg')
img_rgb = cv2.cvtColor(im, cv2.COLOR_BGR2RGB)
text = pytesseract.image_to_string(im, config='-c preserve_interword_spaces=1') 
print(text)
```

# Python Docx for Word Docs

We used docx to extract the clauses, find clauses by line breaks.

```python
import docx

def docx_extraction(docx_path):
    document = docx.Document(docx_path)
    doc_text = '\n\n'.join(paragraph.text for paragraph in document.paragraphs if len(paragraph.text) > 3)
    return doc_text
print(docx_extraction('../reference/sample_eula_1.docx'))
```
